```python
# unpivot cast names
constraints=df['cast'].apply(lambda x:str(x).split(', ')).tolist()
df_new=pd.DataFrame(constraints,index=df['title'])
df_new=df_new.stack()
df_new=pd.DataFrame(df_new)
df_new.reset_index(inplace=True)
df_new=df_new[['title',0]]
df_new.columns=['title','cast']
df_new
#df_new[df_new['cast']]=='Rupa Bhimani'
```

```python
#unpivot director names
constraints_d=df['director'].apply(lambda x:str(x).split(', ')).tolist()
df_newd=pd.DataFrame(constraints_d,index=df['title'])
df_newd=df_newd.stack()
df_newd=pd.DataFrame(df_newd)
df_newd.reset_index(inplace=True)
df_newd=df_newd[['title',0]]
df_newd.columns=['title','director']
df_newd
```

```python
# unpivot country names
constraints_c=df['country'].apply(lambda x:str(x).split(', ')).tolist()
df_newc=pd.DataFrame(constraints_c,index=df['title'])
df_newc=df_newc.stack()
df_newc=pd.DataFrame(df_newc)
df_newc.reset_index(inplace=True)
df_newc=df_newc[['title',0]]
df_newc.columns=['title','country']
df_newc
```

```python
# unpivot listed_in
constraints_g=df['listed_in'].apply(lambda x:str(x).split(', ')).tolist()
df_newg=pd.DataFrame(constraints_g,index=df['title'])
df_newg=df_newg.stack()
df_newg=pd.DataFrame(df_newg)
df_newg.reset_index(inplace=True)
df_newg=df_newg[['title',0]]
df_newg.columns=['title','listed_in']
df_newg
```

```python
# combine the tables using merge
pda=pd.merge(df_new,df_newd,how='inner',on='title')
pdb=pd.merge(df_newc,df_newg,how='inner',on='title')
pdc=pd.merge(pda,pdb,how='inner',on='title')
pd_t=pd.merge(pdc,df,how='inner',on='title')
pd_t=pd_t.drop(columns=['director_y', 'cast_y','country_y','listed_in_y'])
pd_t.rename(columns = {'cast_x':'cast','director_x':'director','country_x':'country','listed_in_x':'listed_in'}, inplace = True)
pd_t
```

```python
country_type= df[['type', 'country']]
def Mov(x):
    if x == "Movie":
        return 1
def Tv(x):
    if x == "TV Show":
        return 1

country_type['movie'] = list(map(Mov, country_type.type))
country_type['tvshow'] = list(map(Tv, country_type.type))
country_type.drop('type', axis=1, inplace=True)
country_type
```

```python
country_type = country_type.groupby('country').sum().sort_values(by='movie', ascending=False)[:10].reindex()
country_type
```

```python
country_type['movie'] = (country_type['movie']*100)/(country_type['movie'] + country_type['tvshow'])
country_type['tvshow'] =(country_type['tvshow']*100)/(country_type['movie'] + country_type['tvshow'])
country_type
```

```python
plt.figure(figsize=(13, 9))
plt.title("movies and tv shows vs country", fontsize=18)
plt.xlabel("countries", fontsize=14)
plt.ylabel("percentage", fontsize=14)
plt.bar(country_type.index, country_type.movie, color='#2f433e')
plt.bar(country_type.index,country_type.tvshow, bottom=country_type.movie, color='#d7b6a1')
plt.legend(['Movies', 'TV Shows'], loc='best')
plt.show()
```
